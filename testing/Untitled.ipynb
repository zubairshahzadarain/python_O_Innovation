```python
import pandas as pd
import cv2
import numpy as np
import pandas as pd
# Load CSV
data = pd.read_csv('./data/img.csv')
#data=data.head(1)
data = data.dropna()
# Resize function
def resize_image(row):
    image = np.array(row[1:]).reshape(1, -1).astype(np.uint8)
    #cv2.imwrite('original_image.jpg', image)
    resized = cv2.resize(image, (150, 1), interpolation=cv2.INTER_CUBIC)
    cv2.imwrite('resized_image.jpg', resized)
    print(f"Colored image size: {resized.shape}")
    return resized ,resized.shape

data[['image_base64', 'image_shape']] = data.iloc[:, 1:].apply(resize_image, axis=1, result_type='expand')
```

```python
def apply_colormap(image_data):
    image = np.frombuffer(image_data, dtype=np.uint8).reshape(1, 150)
    colormap_image = cv2.applyColorMap(image, cv2.COLORMAP_JET)
    return colormap_image.tobytes()
```
